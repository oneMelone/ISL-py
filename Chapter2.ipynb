## Exercises - applied

```python
College = pd.read_csv('College.csv', index_col = 0)
```

```python
College.head()
```

```python
College.describe()
```

```python
College[College['perc.alumni'] > 50]
```

```python
sns.pairplot(College.iloc[:, :10], diag_kind = 'kde', plot_kws = {'alpha': 0.2})
```

```python
College.boxplot(by='Private', column='Outstate',
                           figsize=(5, 5))
```

```python
College['Elite'] = None
```

```python
for index, row in College.iterrows():
    if (row['Top10perc'] <= 50):
        College.loc[index, 'Elite'] = 'No'
    else:
        College.loc[index, 'Elite'] = 'Yes'
```

```python
sns.regplot('Top10perc', 'perc.alumni', data = College)
```

 We can see a relationship between these two parameters.
